### ai_fix_grammar
The ai_fix_grammar() function allows you to invoke a state-of-the-art generative AI model to correct grammatical errors in a given text using SQL. This function uses a chat model serving endpoint made available by Databricks Foundation Model APIs.

Documentation: https://docs.databricks.com/en/sql/language-manual/functions/ai_fix_grammar.html

```python
%sql
SELECT ai_fix_grammar('This sentence have some mistake') as fixed_sentence;
```

### ai_query
The ai_query() function allows you to query machine learning models and large language models served using Mosaic AI Model Serving. To do so, this function invokes an existing Mosaic AI Model Serving endpoint and parses and returns its response. Databricks recommends using ai_query with Model Serving for batch inference

Documentation: https://docs.databricks.com/en/large-language-models/ai-functions.html#ai_query

We can switch models depending on what we are trying to do. See how the performance varies between the 70B model and 8B model below. Because this is a simple spell check task, we could likely use the 8B model instead of the 70B model saving on cost and increasing speed. 

```python
%sql 
SELECT
  `Misspelled_Make`,   -- Placeholder for the input column
  ai_query(
    'databricks-meta-llama-3-1-8b-instruct',
    CONCAT(format_string('You will always receive a make of a car. Check to see if it is misspelled and a real car. Correct the mistake. Only provide the corrected make. Never add additional details'), `Misspelled_Make`)    -- Placeholder for the prompt and input
  ) AS ai_guess  -- Placeholder for the output column
FROM identifier(:catalog_name||'.'||:schema_name||'.'||'synthetic_car_data')
-- limit 3;
```

## ai_parse_document

The ai_parse_document() function invokes a state-of-the-art generative AI model from Databricks Foundation Model APIs to extract structured content from unstructured documents. The following file formats are supported:

- PDF
- JPG / JPEG
- PNG

Documentation: https://docs.databricks.com/aws/en/sql/language-manual/functions/ai_parse_document 

Let's see an example using the image below:

<img src="https://www.databricks.com/sites/default/files/inline-images/blog-marketecture-1.png?v=1700106342" width="50%">

```python
%sql

-- Use the variables in the SQL query
WITH raw_data AS (
  SELECT
    path,
    content
  FROM
    READ_FILES('/Volumes/'||:catalog_name||'/'||:schema_name||'/dip', format => 'binaryFile')
)
SELECT
  path,
  ai_parse_document(content) AS parsed
FROM
  raw_data;
```

```python

%sql
WITH parsed AS (
  SELECT
    path,
    ai_parse_document(content) AS doc
  FROM READ_FILES('/Volumes/agent_bricks_demo/data_schema/dip', format => 'binaryFile')
)
SELECT
  path AS archivo,
  doc:document.elements AS elementos,
  doc:error_status AS estado_error,
  doc:metadata.id AS metadata_id
FROM parsed;
```

### 1. AI_ANALYZE_SENTIMENT

```python
%sql
--Ejemplo 1: Analizar sentimiento de comentarios sobre los modelos
SELECT 
  Correct_Make,
  Correct_Model,
  ai_analyze_sentiment('I love my ' || Correct_Make || ' ' || Correct_Model || ', it runs perfectly!') AS sentimiento_positivo
FROM identifier(:catalog_name||'.'||:schema_name||'.'||'synthetic_car_data') limit 100
```

```python
%sql
-- Ejemplo 2: Analizar sentimiento de quejas
SELECT 
  Correct_Make,
  Correct_Model,
  ai_analyze_sentiment('My ' || Correct_Make || ' ' || Correct_Model || ' keeps breaking down. Worst car ever!') AS sentimiento_negativo
FROM identifier(:catalog_name||'.'||:schema_name||'.'||'synthetic_car_data') limit 100
```

### 2. AI_CLASSIFY

```python
%sql
-- Ejemplo 1: Clasificar tipo de vehculo
SELECT 
  Correct_Make,
  Correct_Model,
  ai_classify(
    Correct_Model,
    ARRAY('Sedan', 'Truck', 'SUV', 'Sports Car')
  ) AS tipo_vehiculo
FROM identifier(:catalog_name||'.'||:schema_name||'.'||'synthetic_car_data') limit 100
```

```python
%sql
-- Ejemplo 2: Clasificar nivel de error ortogrfico
SELECT 
  Misspelled_Make,
  Misspelled_Model,
  ai_classify(
    CONCAT('Make: ', Misspelled_Make, ', Model: ', Misspelled_Model),
    ARRAY('Error Leve', 'Error Moderado', 'Error Grave')
  ) AS nivel_error
FROM identifier(:catalog_name||'.'||:schema_name||'.'||'synthetic_car_data') limit 100
```

### 3. AI_EXTRACT

### 6. AI_MASK

```python
%sql
-- Ejemplo 1: Enmascarar informacin sensible especificando qu tipos de datos proteger
SELECT 
  Correct_Make,
  Correct_Model,
  ai_mask(
    'Juan Prez compr un ' || Correct_Make || ' ' || Correct_Model || 
    ' con VIN 1HGBH41JXMN109186 y lo registr con email juan.perez@email.com',
    ARRAY('PERSON', 'EMAIL', 'VIN')  -- Especifica qu enmascarar
  ) AS info_enmascarada
FROM identifier(:catalog_name||'.'||:schema_name||'.'||'synthetic_car_data') 
LIMIT 100;
```

```python
%sql
-- Ejemplo2: Solo enmascarar emails
SELECT 
  Correct_Make,
  ai_mask(
    'Contacto: juan.perez@email.com, ventas@' || LOWER(Correct_Make) || '.com',
    ARRAY('EMAIL')
  ) AS emails_protegidos
FROM identifier(:catalog_name||'.'||:schema_name||'.'||'synthetic_car_data') 
LIMIT 10;
```

```python
%sql
-- Ver original y enmascarado juntos para validar
SELECT 
  Correct_Make,
  'Contacto: juan.perez@email.com, ventas@' || LOWER(Correct_Make) || '.com' AS email_original,
  ai_mask(
    'Contacto: juan.perez@email.com, ventas@' || LOWER(Correct_Make) || '.com',
    ARRAY('EMAIL')
  ) AS email_enmascarado
FROM identifier(:catalog_name||'.'||:schema_name||'.'||'synthetic_car_data') 
LIMIT 10;
```

```python
%sql
-- Max 20 camposs
WITH datos_prueba AS (
  SELECT 
    'Juan Prez (DNI: 12345678-X, Pasaporte: ABC123456) vive en Av. Arequipa 1234, Lima.
     Telfono: +51 987-654-321, Email: juan@example.com
     Tarjeta: 4532-1234-5678-9010, Cuenta: ES91 2100 0418 4502 0005 1332
     Fecha nacimiento: 15/03/1985, SSN: 123-45-6789
     IP: 192.168.1.1, Licencia conducir: B12345678
     VIN: 1HGBH41JXMN109186, Placa: ABC-123' AS texto_original
)
SELECT 
  'ORIGINAL' AS tipo,
  texto_original AS contenido
FROM datos_prueba

UNION ALL

SELECT 
  'ENMASCARADO' AS tipo,
  ai_mask(
    texto_original,
    ARRAY(
      'PERSON', 'DATE', 'EMAIL', 'PHONE', 'ADDRESS',
      'ID_NUMBER', 'PASSPORT', 'SSN', 'DRIVER_LICENSE', 'LICENSE_PLATE',
      'CREDIT_CARD', 'BANK_ACCOUNT', 'IBAN', 'VIN', 'IP_ADDRESS',
      'LOCATION', 'ORGANIZATION', 'URL', 'MAC_ADDRESS', 'AGE'
    )
  ) AS contenido
FROM datos_prueba;
```

```python
%sql
-- Crear una funcin de enmascaramiento reversible (requiere Unity Catalog)
CREATE FUNCTION IF NOT EXISTS mask_email(email STRING)
RETURNS STRING
RETURN 
  CASE 
    WHEN IS_ACCOUNT_GROUP_MEMBER('admins') THEN email  -- Admins ven el real
    ELSE CONCAT(LEFT(email, 3), '***@***.com')        -- Otros ven enmascarado
  END;

-- Crear vista con enmascaramiento dinmico
CREATE OR REPLACE VIEW cars_masked AS
SELECT 
  Correct_Make,
  Correct_Model,
  mask_email('juan.perez@email.com') AS email_contacto
FROM identifier(:catalog_name||'.'||:schema_name||'.'||'synthetic_car_data');

-- Los usuarios ven segn sus permisos
SELECT * FROM cars_masked;
```

### 7. AI_SIMILARITY

```python
%sql
-- Ejemplo 1: Calcular similitud entre nombre mal escrito y correcto
SELECT 
  Misspelled_Make,
  Correct_Make,
  ai_similarity(Misspelled_Make, Correct_Make) AS similitud_marca
FROM identifier(:catalog_name||'.'||:schema_name||'.'||'synthetic_car_data') limit 100
```

```python
%sql
-- Ejemplo 2: Calcular similitud entre modelos
SELECT 
  Misspelled_Model,
  Correct_Model,
  ai_similarity(Misspelled_Model, Correct_Model) AS similitud_modelo
FROM identifier(:catalog_name||'.'||:schema_name||'.'||'synthetic_car_data') limit 100
```

### 8. AI_SUMMARIZE

```python
%sql
-- Ejemplo 1: Resumir resea larga del vehculo
SELECT 
  Correct_Make,
  Correct_Model,
  ai_summarize(
    'El ' || Correct_Make || ' ' || Correct_Model || 
    ' es un vehculo excepcional que combina rendimiento, comodidad y eficiencia. ' ||
    'Su motor es potente y confiable, el interior es espacioso y lujoso, ' ||
    'y la tecnologa integrada es de ltima generacin. ' ||
    'Los propietarios reportan alta satisfaccin y bajos costos de mantenimiento.'
  ) AS resumen_corto
FROM identifier(:catalog_name||'.'||:schema_name||'.'||'synthetic_car_data') limit 100
```

```python
%sql
-- Ejemplo 2: Resumir historial de servicio
SELECT 
  Correct_Make,
  Correct_Model,
  ai_summarize(
    'Historial de servicio del ' || Correct_Make || ' ' || Correct_Model || ': ' ||
    'Cambio de aceite a los 5,000 km. Rotacin de llantas a los 10,000 km. ' ||
    'Cambio de filtro de aire a los 15,000 km. Inspeccin de frenos a los 20,000 km. ' ||
    'Todos los servicios completados sin problemas.'
  ) AS resumen_servicio
FROM identifier(:catalog_name||'.'||:schema_name||'.'||'synthetic_car_data') limit 100
```
